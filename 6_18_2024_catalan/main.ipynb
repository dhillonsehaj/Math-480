## Imports
The files you need for this notebook are imported in the cell below.

```python
import catalan
import os
import time
import matplotlib.pyplot as plt
dirname = os.path.dirname("./")
```

## Generate Parenthesizations

Generate all possible parenthesizations of length $n$ for $1 \le n \le 10$. Files should be outputted to `data/catalan_parenthesizations_n.txt` and a plot of the time used to generate each data file should go in `figures/catalan_parenthesizations.png`.

```python
file_path = os.path.join(dirname, "data/catalan_parenthesizations_n.txt")
with open(file_path, 'w') as file:
  file.write("")
timeplot = []

for n in range(1, 11):
  start = time.time()
  with open(file_path, 'a') as file:
    file.write(str(catalan.parenthesizations(n)) + '\n')
  end = time.time()
  timeplot.append(end - start)

plt.figure()
plt.plot(range(1, 11), timeplot, 'r', label='time')
plt.xlabel('n')
plt.xticks(range(1, 11))
plt.ylabel('time (s)')
plt.title('time vs n lengths for parenthesizations')
plt.savefig(os.path.join(dirname, "figures/catalan_parenthesizations.png"))
plt.show()
```
